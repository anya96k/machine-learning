```python

import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[11,12,30,25,30]
plt.plot(x,y)

plt.title('simple line graph')
plt.xlabel('values of x')
plt.ylabel('values of y')

plt.show()
```

```python
plt.scatter(x,y)
plt.title('scatter plot')
plt.xlabel('values of x')
plt.ylabel('values of y')

plt.show()
```
