```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
a= np.array([1,2,3,4,5,6]).reshape(-1,1) 
b=([10,20,30,40,50,60])
model=LinearRegression()
model.fit(a,b)
beta_0 =model.intercept_
beta_1 =model.coef_[0]
print(f"intercept(beta_0):{beta_1}")
print(f"slope(beta_0):{beta_1}")
b_pred=model.predict(a)
plt.scatter(a,b,color='red',label='actual score')
plt.plot(a,b_pred,color='blue',label='regression line')
plt.xlabel('student')
plt.ylabel('attendence')
plt.legend()
plt.show()
```
